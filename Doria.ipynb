### Who donated the most in number of times

```python
donors.Nome_do_Doador.value_counts().reset_index()
```

The politician João Doria is a millionaire entrepreneur, so he was the one who gave more for his own campaign. Second is Sonia Regina Hess Souza (fashion business owner who sold her company by millions in 2013).

### Filter and add the two - Values in Brazilian currency (real)

```python
first_place = donors[donors.Nome_do_Doador == "JOÃO AGRIPINO DA COSTA DORIA JUNIOR"]
```

```python
second_place = donors[donors.Nome_do_Doador == "SÔNIA REGINA HESS SOUZA"]
```

```python
first_place.Valor.sum()
```

```python
second_place.Valor.sum()
```

### Sort donors by value

```python
donors.sort_values("Valor", ascending=False)
```

Here in Brazil the donations can be made by the party directions as well. So we see that the National Directorate of the party and Doria himself were the biggest donors.

After them the biggest donor was RUBENS O S MELLO.

### Filter and aggregate top donors

```python
donor_Direcao = donors[donors.Nome_do_Doador == "Direção Nacional"]
```

```python
donor_Direcao.Valor.sum()
```

```python
donor_Rubens = donors[donors.Nome_do_Doador == "RUBENS O S MELLO"]
```

```python
donor_Rubens.Valor.sum()
```

His full name is Rubens Ometto Silveira Mello, industry entrepreneur fuels in Brazil.

### Total sum of all donors, in order of value

```python
donors.groupby("Nome_do_Doador").Valor.sum().reset_index().sort_values("Valor",ascending=False)
```

### Charts from major donors

```python
%matplotlib inline
```

```python
top_supporters = donors.groupby(
    ["Nome_do_Doador"]
).Valor.sum().reset_index().sort_values("Valor", ascending=False).head(10)
```

```python
top_supporters.head(10)
```

```python
top_supporters.Valor.plot.bar()
```

### With the names

```python
chart = top_supporters.head(10).Valor.plot.barh()
chart.set_yticklabels(top_supporters.Nome_do_Doador)
```
